**Intersection network**

*M/M/1 queue model*

The following intersection network consists of two four-way intersections in a 1x2 grid, each 30 m wide and 250 m apart. Each vehicle moves 14 m/s (50.4 km/h) at full speed.

```python
intersection_system_1 = Model1.IntersectionNetworkSimulator()
intersection_system_1.initialize(grid_dimensions=(3,3), grid_distance=250)

intersection_system_1.set_queue_rate_parameters(grid_ind=(0,0), avg_departure_time=1, avg_arrival_time_s=10, avg_arrival_time_e=10)
intersection_system_1.set_queue_rate_parameters(grid_ind=(0,1), avg_departure_time=1, avg_arrival_time_s=10)
intersection_system_1.set_queue_rate_parameters(grid_ind=(0,2), avg_departure_time=1, avg_arrival_time_s=10, avg_arrival_time_w=10)
intersection_system_1.set_queue_rate_parameters(grid_ind=(1,0), avg_departure_time=1, avg_arrival_time_e=10)
intersection_system_1.set_queue_rate_parameters(grid_ind=(1,1), avg_departure_time=1)
intersection_system_1.set_queue_rate_parameters(grid_ind=(1,2), avg_departure_time=1, avg_arrival_time_w=10)
intersection_system_1.set_queue_rate_parameters(grid_ind=(2,0), avg_departure_time=1, avg_arrival_time_n=10, avg_arrival_time_e=10)
intersection_system_1.set_queue_rate_parameters(grid_ind=(2,1), avg_departure_time=1, avg_arrival_time_n=10)
intersection_system_1.set_queue_rate_parameters(grid_ind=(2,2), avg_departure_time=1, avg_arrival_time_n=10, avg_arrival_time_w=10)

intersection_system_1.set_traffic_lights(grid_ind=(0,0), traffic_light_ns=periodic_ns11, traffic_light_ew=periodic_ew11)
intersection_system_1.set_traffic_lights(grid_ind=(0,1), traffic_light_ns=periodic_ns12, traffic_light_ew=periodic_ew12)
intersection_system_1.set_traffic_lights(grid_ind=(0,2), traffic_light_ns=periodic_ns13, traffic_light_ew=periodic_ew13)
intersection_system_1.set_traffic_lights(grid_ind=(1,0), traffic_light_ns=periodic_ns21, traffic_light_ew=periodic_ew21)
intersection_system_1.set_traffic_lights(grid_ind=(1,1), traffic_light_ns=periodic_ns22, traffic_light_ew=periodic_ew22)
intersection_system_1.set_traffic_lights(grid_ind=(1,2), traffic_light_ns=periodic_ns23, traffic_light_ew=periodic_ew23)
intersection_system_1.set_traffic_lights(grid_ind=(2,0), traffic_light_ns=periodic_ns31, traffic_light_ew=periodic_ew31)
intersection_system_1.set_traffic_lights(grid_ind=(2,1), traffic_light_ns=periodic_ns32, traffic_light_ew=periodic_ew32)
intersection_system_1.set_traffic_lights(grid_ind=(2,2), traffic_light_ns=periodic_ns33, traffic_light_ew=periodic_ew33)
```

```python
intersection_system_1.set_observable_intersections(grid_inds=intersection_system_1.grid_inds)
```

```python
intersection_system_1.simulate(delta_t=0.1, end_time=120, animate=True, fig_width=4, file_name="simulation1_3x3.mp4", speed=3)
```

View the video below.

```python
fig, axs = intersection_system_1.plot_queue_stats(plt=plt, grid_ind=(0,1), direction='E', end_time=round(intersection_system_1.time), delta_t=0.1, traffic_light=periodic_ew12,fig_size=(6,6))
plt.savefig(output_path+'figs/3x3_eastbound_(0,1)_1', bbox_inches='tight')
```
